```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from collections import namedtuple
plt.style.use('fivethirtyeight')
```

```python
n_groups = 4

means_men = (694.8295589437141, 347.15174185620157, 300.2988351463691, 159.97215954810648)

means_women = (585.6584321621915, 347.15174185620157, 200.18207572026213, 168.26441100954486)

fig, ax = plt.subplots(figsize=(14,6))

index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = ax.bar(index, means_men, bar_width,
                alpha=opacity, color='b',
                error_kw=error_config,
                label='With NIFTY')

rects2 = ax.bar(index + bar_width, means_women, bar_width,
                alpha=opacity, color='r',
                error_kw=error_config,
                label='Without NIFTY')

plt.figure(figsize=(14,6))
ax.set_xlabel('Models')
ax.set_ylabel('RMSE')
ax.set_title('TCS RMSE')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(('VAR', 'XGBOOST', 'LSTM', 'GRU'))
ax.legend(loc = 'best')

fig.tight_layout()
plt.show()
```
